# Matplotlib 연습장
데이터 시각화를 위한 기본적인 matplotlib 예제를 알아봅시다.

## 1. Matplotlib 사용하기
소스코드 최상단에 아래와 같이 입력합니다. 보통 plt를 약자로 사용

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 2. 기본 그래프 그리기
plot() 함수를 이용해 대부분의 그림을 그릴 수 있다.


```python
x = np.arange(0,10,1)
x
```

```python
y = np.random.randn(10)
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
y1 = np.random.randn(10)
y2 = np.random.randn(10)

plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

## 3. 그래프 옵션 조정
1. plot() 옵션  
- color : 선 색깔 변경
- marker : 데이터 꺽임 부분 표시
- label : 범례 지정
- linestyle : 선 모양

2. 그래프 옵션   
- plt.xlabel() : x축 설명
- plt.ylabel() : y축 설명
- plt.title() : 그래프 제목
- legend() : 범례 표시
- plt.xlim() : x축 범위
- plt.ylim() : y축 범위

```python
plt.plot(x,y1, color = "red", marker=".")
plt.plot(x,y2, color = "black", marker = "*", linestyle = "--")
plt.show()
```
