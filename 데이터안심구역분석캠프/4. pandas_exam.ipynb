#### 고객들이 직업별로 구매금액이 어떻게 될까요?

```python
df_tmp = df[['Job','Purchase Price']]
df_tmp
```

```python
df_tmp.groupby('Job') #바로 볼 수 없음
```

```python
df_tmp.groupby('Job').count() #구매 금액 건수
```

```python
df_tmp.groupby('Job').mean()
```

```python
df_tmp.groupby('Job').max()
```

#### 어떤 직업을 가진 사람들이 가장 많은 금액을 지출했나요?

```python
df_tmp.groupby('Job').sum()
```

```python
df_tmp.groupby('Job').sum().sort_values(by='Purchase Price')
```
